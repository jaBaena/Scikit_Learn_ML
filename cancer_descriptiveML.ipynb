#### K-vecinos.

Primero hacemos los importes necesarios y definimos la matriz de confusión:

```python
from sklearn.utils.multiclass import unique_labels
import warnings
import numpy as np
warnings.filterwarnings('ignore')

def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = classes[unique_labels(y_true, y_pred)]
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

Para llevar a cabo este algoritmo, primero se divide el dataset en dos partes, una para entrenamiento y otra para testear el resultado.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 13)
```

Antes de realizar una busqueda entre los hiperparámetros de este método, se calculará una estimación del valor de k en función de la precisión que se puede obtener con la validación cruzada. De esta manera se obtiene un rango de K en el que la precisión de la validación cruzada obtiene una precisión aceptable. La validación cruzada dividirá la muestra en  partes distintas.

```python
from sklearn.metrics import accuracy_score 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

# creating odd list of K for KNN
myList = list(range(1,50))

# subsetting just the odd ones
neighbors = myList

# empty list that will hold cv scores
cv_scores = []

# perform 10-fold cross validation
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())

# changing to misclassification error
MSE = [1 - x for x in cv_scores]

# plot misclassification error vs k
plt.plot(neighbors, MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```

Una vez estimado los posibles valores de K, se define los parámetros y sus valores entre los que se buscará el modelo con una preción más alta. Los parámetros serán los siguientes

- 'n_neighbors': número de vecinos que definirán la clase de cada muestra.
- 'p': Parámetro para la métrica Minkowski. Si es igual a 1, se usará la métrica Manhattan, si es igual a 2 una métrica euclídea y para el resto de valores, distintas métricas Minkowski. (Con estos valores se evita definir el parámetro 'metric')

A continuación se mostrarán los parámetros que definen el mejor modelo de entre todos los posibles.

```python
from sklearn.neighbors import KNeighborsClassifier

param_grid = {'n_neighbors':[1,3,5,7,9,11,13], 'p':[2,3,4,5,6,7,8,9]}

from sklearn.model_selection import GridSearchCV, KFold

miKNN = KNeighborsClassifier()
miGSCV = GridSearchCV(estimator=miKNN,param_grid=param_grid,scoring='accuracy',cv=KFold(n_splits=10,random_state=3))

miGSCV.fit(X_train, y_train)

miGSCV.best_estimator_.fit(X_train, y_train)

y_pred = miGSCV.predict(X_test)
```

Una vez se ha realizado el análisis obtenemos la matriz de confusión, donde se puede ver el error de clasificación de cada clase:

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,y_pred))

plot_confusion_matrix(y_test, y_pred, classes=data['target_names'], normalize=True,
                      title='Matriz de confusión')
```

### 5) Razona si es necesario estratificar el proceso de cross-validation analizando la distribución de muestras por clase.

La estratifcación del proceso de validación cruzada es necesario cuando la distribución de los datos no es uniforme. Si en el dataset hay más instancias de una clase que de otra, las muestras que se realizan durante la validación cruzada no tienen porqué ser representativas del dataset. Pueden aparacer muestras sesgadas que no sean representativas, induciendo a error al modelo.

Para ello, vemos la proporción de cada etiquetas de cada clase.

```python
import pandas as pd

unique, counts = np.unique(y, return_counts=True)
k=np.asarray((unique, counts)).T

df = pd.DataFrame({'label':k[:,0],'count':k[:,1]})
df['%']=(df['count']/df['count'].sum()).round(2)
df
```

Como se puede ver, la distribución de las etiquetas no es uniforme. A continuación se vuelve a realizar el método K-vecinos pero añadiendo la estratificación a la validación cruzada.

```python
from sklearn.model_selection import StratifiedKFold

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state = 13)

param_grid = {'n_neighbors':[1,3,5,7,9,11,13], 'p':[2,3,4,5,6,7,8,9]}

miGSCV3 = GridSearchCV(estimator=miKNN,param_grid=param_grid,scoring='accuracy',cv=StratifiedKFold(n_splits=10,random_state=3))

miGSCV3.fit(x_train, y_train)

miGSCV3.best_estimator_.fit(x_train, y_train)

y_pred3 = miGSCV3.predict(x_test)

print(confusion_matrix(y_test,y_pred3))

plot_confusion_matrix(y_test, y_pred3, classes=data['target_names'], normalize=True,
                      title='Matriz de confusión')
```

Como se puede ver, despues de estratificar las muestras, ha aumentado la efectividad.
